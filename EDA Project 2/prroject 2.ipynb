Introduction to the Global Finance Dataset

The Global Finance dataset provides a comprehensive view of worldwide financial indicators and economic performance across different countries and regions. It is designed to capture essential aspects of the global economy by integrating data on financial stability, investment trends, trade balances, and development indicators.

This dataset typically includes information such as:
.Country/Region – the geographical scope of the data
.GDP and Growth Rates – measures of economic performance
.Inflation and Interest Rates – indicators of price stability and monetary policy
.Foreign Direct Investment (FDI) – flows of international capital
.Trade Balances – exports and imports for each country
.Unemployment Rates – labor market conditions
.Stock Market Indices / Currency Exchange Rates – financial market performance

Importance of the Dataset

.Global Economic Analysis – Helps researchers and analysts study trends in global growth, inflation, and trade.
.Policy Making – Governments and international bodies use such data to assess economic stability and plan interventions.
.Investment Decisions – Financial institutions rely on these indicators to identify opportunities and risks in different regions.
.Comparative Studies – Enables comparisons of economic performance between developed, emerging, and underdeveloped economies.

Applications
.Forecasting future financial and economic trends
.Identifying regional strengths and weaknesses
.Supporting academic research in economics and finance
.Assisting businesses in market expansion strategies

Outlier Detection

```python
def remove_outliers_iqr(data, columns):
    df_clean = data.copy()
    for col in columns:
        if pd.api.types.is_numeric_dtype(df_clean[col]):  # Only numeric columns
            Q1 = df_clean[col].quantile(0.25)
            Q3 = df_clean[col].quantile(0.75)
            IQR = Q3 - Q1
            lower = Q1 - 1.5 * IQR
            upper = Q3 + 1.5 * IQR

            print(f"\nColumn: {col}")
            print(f"Lower bound: {lower}, Upper bound: {upper}")
            print(f"Outliers detected: {((df_clean[col] < lower) | (df_clean[col] > upper)).sum()}")

            # Remove outliers
            df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]
    return df_clean

# Apply on all numeric columns
numeric_cols = df.select_dtypes(include="number").columns
df_clean = remove_outliers_iqr(df, numeric_cols)

print("\nOriginal shape:", df.shape)
print("Cleaned shape:", df_clean.shape) 




# Before removing outliers
plt.figure(figsize=(12,6))
sns.boxplot(data=df[numeric_cols])
plt.title("Before Removing Outliers")
plt.show()

# After removing outliers
plt.figure(figsize=(12,6))
sns.boxplot(data=df_clean[numeric_cols])
plt.title("After Removing Outliers")
plt.show()
```

General Suggestions & Recommendations for Global Finance Dataset
🔹 1. Data Cleaning

Handle missing values → Check for nulls (df.isnull().sum()) and decide whether to fill (mean/median) or drop.

Outlier detection → Use IQR or Z-score methods to remove extreme values (as we discussed).

Standardize units → Ensure values like GDP are consistently in trillions/billions (avoid mixing currencies).

🔹 2. Exploratory Data Analysis (EDA)

Summary stats → df.describe() for central tendency and spread.

Distribution checks → Histograms/boxplots to see skewness in GDP, Inflation, FDI, etc.

Trends over time → Line plots for GDP, inflation, unemployment, etc.

Top vs bottom performers → Identify best and worst countries for each metric.

🔹 3. Visualization Recommendations

Line charts → GDP trends for major economies (India, USA, China).

Bar charts → Compare average GDP across countries.

Heatmaps → Show correlations (e.g., GDP vs FDI, Inflation vs Unemployment).

Scatterplots → Explore relationships (e.g., GDP per Capita vs Inflation).

🔹 4. Analytical Insights to Explore

GDP vs Inflation → Does higher inflation hurt growth?

Exports vs Imports (Trade Balance) → Which countries are net exporters/importers?

FDI inflows → Which countries attract the most investment?

Unemployment trends → Global vs country-specific.

GDP per Capita vs Quality of life → Proxy indicator of economic development.

🔹 5. Recommendations for Further Analysis

Normalize per capita → Use GDP per Capita instead of raw GDP for fair comparisons.

Regional grouping → Compare continents/regions (e.g., Asia vs Europe).

Time-series forecasting → Apply models (ARIMA, Prophet) to forecast GDP/Inflation.

Cluster analysis → Group countries with similar economic profiles.

Policy insights → Countries with low GDP but high FDI could focus on export growth.
