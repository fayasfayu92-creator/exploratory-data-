Introduction


This dataset presents valuable information about the water quality of rivers across Jharkhand, India.
It contains measurements recorded at different monitoring locations during the year 2022.Each entry is identified by a unique station code assigned to the monitoring site.The dataset captures environmental parameters of river water at multiple dams and river stretches.
It includes data from rivers like Jumar, Ajay, Konar, and others located across Ranchi, Bokaro, and nearby districts.The type of water body in all records is identified as a river.
Key physical characteristics such as minimum and maximum water temperature are documented.Dissolved oxygen levels, an essential indicator of aquatic health, are also measured.pH values provide insights into the acidity or alkalinity of the water samples.Electrical conductivity is included to indicate the presence of dissolved salts and ions.Biochemical Oxygen Demand (BOD) values highlight the organic pollution levels in the rivers.In addition, Nitrate-N concentrations are reported where available.
The dataset also records microbiological indicators like fecal coliform and total coliform counts.
These parameters are crucial in determining the suitability of river water for human and ecological use.The records show variations in water quality across different sites and seasons.Such variations are influenced by natural factors and human activities around the rivers.
The data provides a basis for environmental monitoring and policy planning.
It can help identify pollution hotspots and regions with relatively good water quality.
Researchers, policymakers, and environmentalists can use this dataset for deeper analysis.
Overall, the dataset contributes to understanding the state of river ecosystems in Jharkhand

```python
# Function to detect and remove outliers using IQR
def remove_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Detect outliers
    outliers = data[(data[col] < lower) | (data[col] > upper)]
    print(f"\nColumn: {col} | Outliers detected: {len(outliers)}")
    
    # Plot before removing outliers
    plt.figure(figsize=(6,4))
    data.boxplot(column=col)
    plt.title(f"Before Removing Outliers - {col}")
    plt.show()
    
    # Remove outliers
    cleaned_data = data[(data[col] >= lower) & (data[col] <= upper)]
    
    # Plot after removing outliers
    plt.figure(figsize=(6,4))
    cleaned_data.boxplot(column=col)
    plt.title(f"After Removing Outliers - {col}")
    plt.show()
    
    return cleaned_data
df_clean = df.copy()
for col in numeric_cols:
    df_clean = remove_outliers_iqr(df_clean, col)

print("\nFinal cleaned dataset shape:", df_clean.shape)

```

🔹 General Suggestions & Recommendations for Indian Water Data
1. Water Quality Parameters

pH should ideally be between 6.5–8.5.

If below 6.5 → water is acidic, corrosive to pipes.

If above 8.5 → alkaline, may cause scaling.
Recommendation: Neutralization treatment (lime for acidic water, aeration for alkaline).

TDS (Total Dissolved Solids) should be < 500 mg/L (BIS/WHO standard).

High TDS → salty/bitter taste, harmful for kidney patients.
Recommendation: Use RO purification if TDS > 500.

Hardness (Calcium & Magnesium salts) should be < 300 mg/L.

Very hard water → scaling in pipes, affects soaps/detergents.
Recommendation: Use water softeners for households/industries.

Chlorides should be < 250 mg/L.

Excess chloride → salty taste, corrosion of pipes.
Recommendation: Desalination (if coastal groundwater is used).

2. Geographical Trends

Certain states may naturally have higher TDS (e.g., Rajasthan, Gujarat).

States with industrial areas may show higher pollutants.
Recommendation: Focus on state-wise treatment solutions rather than one-size-fits-all.

3. Public Health

Unsafe pH, high TDS, and hardness can cause digestive, skin, and kidney issues.

Long-term exposure to high fluoride, nitrate, or arsenic can cause serious health risks.
Recommendation: Continuous monitoring + awareness campaigns in rural areas.

4. Data Utilization

Create heatmaps and trend graphs to visualize which states have the worst water quality.

Identify the top 5 states with the best and worst quality for targeted policy actions.

5. Policy-Level Recommendations

Invest in community RO plants for rural areas with high TDS.

Strengthen groundwater monitoring networks.

Promote rainwater harvesting to dilute contaminants.

Ensure strict industrial waste disposal regulations.
